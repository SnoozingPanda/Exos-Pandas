# Your mission: convert the datetime object into string with precise format

```
var_date = datetime.strptime('2020-02-15 20:05', '%Y-%m-%d %H:%M')
print(var_date)
```

```
# We would like to print this format : 15/02/2020
date1 = datetime.strftime(var_date, "%d/%m/%Y")
print(date1)

```

```
# We would like to print this format : 15-02-2020
date2 = datetime.strftime(var_date, "%d-%m-%Y")
print(date2)

```

```
# We would like to print this format : 15/02
date3 = datetime.strftime(var_date, "%d/%m")
print(date3)

```

```
# We would like to print this format : 2020-02
date4 = datetime.strftime(var_date, "%Y-%m")
print(date4)

```

```
# We would like to print this format : 2020-02-15 at 20:05
date5 = datetime.strftime(var_date, "%Y-%m-%d at %H:%M")
print(date5)

```

# Your mission is to determine which date it was / it will be

```
# what month will it be in 7 weeks ?
date_future = datetime.now() + timedelta(weeks = 7)
mois = datetime.strftime(date_future, "%m")
print(mois)
```

```
# what day of the week will it be in 3 days ?
date_future1 = datetime.now() + timedelta(days = 3)
day = datetime.strftime(date_future1, "%d")
print(day)

```

```
# what hour will it be in 100 hours ?
heure_future = datetime.now() + timedelta(hours = 11)
hour = datetime.strftime(heure_future, "%H")
print(hour)

```

```
# what date was 30 days ago ?
date_hier = datetime.now() - timedelta(days = 30)
hier = datetime.strftime(date_hier, "%d-%m-%Y")
print(hier)

```

```
# what date was 30 weeks ago ?
date_avt_hier = datetime.now() - timedelta(weeks = 30)
avt_hier = datetime.strftime(date_avt_hier, "%d-%m-%Y")
print(avt_hier)

```
