#### Deciles

Here you can find [a little help](https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.quantile.html)

```
# Find a way to print the 9 deciles + min + max of the column WEATHER_CODE_EVENING with the quantile method of pandas and a loop

list_quantiles = []
for i in range(1, 10):
  decile = df2018["WEATHER_CODE_EVENING"].quantile(i / 10)
  list_quantiles.append(decile)

min_weather = df2018["WEATHER_CODE_EVENING"].min()
max_weather = df2018["WEATHER_CODE_EVENING"].max()

print("decile:")
for i, value in enumerate(list_quantiles):
  print(f"decile {i+1}: {value}")

print(f"Min: {min_weather}")
print(f"Max: {max_weather}")
```

```
#Crée une fonction translate qui traduit chacune des 5 valeurs possibles présente dans OPINION en français.
#Ta fonction doit prendre un str en paramètre, et retourner un autre str avec la traduction.
#Par exemple : translate("bad") doit retourner "mauvais". Idem pour very bad, not good not bad, good et very good.
#Tu peux utiliser un dictionnaire dict, ou bien des conditions imbriquées (le dictionnaire est plus élégant).
#Attention, tu dois prévoir le cas où l'argument fourni n'est pas l'une des 5 valeurs, et retourner alors le texte inchangé.
#Par exemple : translate("hello") doit retourner "hello".
#Applique avec apply ta fonction à la Série OPINION, vérifie que ca fonctionne, puis stocke le résultat dans cette même colonne OPINION.
```

```
dico = {"bad": "mauvais",
          "very bad": "très mauvais",
          "not good not bad": "ni bon ni mauvais",
          "good": "bon",
          "very good": "très bon"}

def translate(str):
  for valeur in dico.keys():
    if str == valeur:
      return dico[str]
  return "inconnu"


df2018["OPINION"] = df2018["OPINION"].apply(translate)
```
